```python
from rdkit.Chem.rdMolAlign import CalcRMS 
from rdkit.Chem.rdMolAlign import AlignMolConformers
from rdkit import Chem
from rdkit.Chem import AllChem
from pathlib import Path
from rdkit.Chem.rdMolAlign import AlignMol
import pandas as pd
```

```python
HERE = Path(_dh[-1])
```

```python
# 导入spyrmsd模块
from spyrmsd import io, rmsd
```

```python
DATA = HERE /"3fqk"
df3fqk = pd.DataFrame(columns=['molecule1', 'molecule2', 'rmsd'])
l=0
for i in range(1,83):
    
    for j in range(1,10):
       
        try:
            file1 = HERE / f"palm2_initial/palm2_{i}.pdb"
            ref = io.loadmol(file1)
            file2 = DATA / f"palm2_{i}_{j}.pdb"
            dock = io.loadmol(file2)
            ref.strip()
            dock.strip()
            # 获取原子坐标、原子序数和邻接矩阵
            coords_ref = ref.coordinates
            anum_ref = ref.atomicnums
            adj_ref = ref.adjacency_matrix
            coords_dock = dock.coordinates
            anum_dock = dock.atomicnums
            adj_dock = dock.adjacency_matrix
            # 计算对称校正的RMSD（可选择是否最小化）
            rmsd_value = rmsd.symmrmsd(
                coords_ref,
                coords_dock,
                anum_ref,
                anum_dock,
                adj_ref,
                adj_dock,
                minimize=True # 或者 False
            )
            # 打印RMSD值
            print(rmsd_value)
        except:
            print(f"--------------------------------------{i}_{j}------------------------ ")
```

```python
DATA = HERE /"3fqk"
df3fqk = pd.DataFrame(columns=['molecule1', 'molecule2', 'rmsd'])
l=0
for i in range(1,127):
    
    for j in range(1,10):
       
        try:
            file1 = HERE / f"palm2_initial/palm2_{i}.pdb"
            print(file1)
            file2 = DATA / f"palm2_{i}_{j}.pdb"
            print(file2)
            mol1 = Chem.MolFromPDBFile(str(file1))
            #mol1 = AllChem.AssignBondOrdersFromTemplate(ref, mol1)
            mol2 = Chem.MolFromPDBFile(str(file2))
            mol2 = AllChem.AssignBondOrdersFromTemplate(mol1, mol2)
            rms = AllChem.AlignMol(mol1, mol2)
            print(rms,i,j)
            df3fqk.loc[l] = [f"{i}_0",f"{i}_{j}",rms]
            l = l+1
        except:
            print(f"--------------------------------------{i}_{j}-------------------------------------------")
```

```python
df3fqk.to_csv('./rmsd_3fqk.csv', mode='a', index=False, header=False)
```
